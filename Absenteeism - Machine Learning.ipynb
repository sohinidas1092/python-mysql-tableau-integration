## STANDARDIZE THE DATA

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
# This is a custom scaler based on standard scaler from sklearn.
# Custom scaler will not standardize all inputs but only the ones we choose.In this way we preserve the dummies untouched.
# We choose the columns to be scaled.
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator,TransformerMixin): 
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        self.scaler = StandardScaler(copy,with_mean,with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None

    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self

    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

```python
#columns_to_scale = ['Month Value','Day of the Week', 'Transportation Expense', 'Distance to Work',
       #'Age', 'Daily Work Load Average', 'Body Mass Index', 'Children', 'Pet']
# We omit the dummy variables from the list.
columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Education']
```

```python
# We create a new variable called columns_to_scale that contains the features to be scaled.
columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_omit]
```

```python
from sklearn.model_selection import train_test_split
# It splits the arrays and matrices into test and train subsets.
```

## SPLIT

```python
# Our inputs are contained into 'scaled_inputs' and our targets are contained into 'targets'.
# It gives a training dataset with inputs and targets and a test dataset with inputs and targets with 4 arrays.
# If we rerun our code we get a different split and shuffle each time. It can can higher as well as lower accuracy with shuffling.
train_test_split(unscaled_inputs, targets)
```

```python
# We now create a summary table containing feature name information.
summary_table  = pd.DataFrame(columns = ['Feature_name'], data = feature_name)
# To create a new column called coefficient
summary_table ['Coefficient'] = np.transpose(reg.coef_)
```

```python
summary_table
```

```python
# To add the intercept
# To shift up all indices by 1
summary_table.index = summary_table.index+1
```

```python
# Now the 0th row is empty. We fill it with the intercept.
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
```

```python
summary_table = summary_table.sort_index()
summary_table
```

## INTERPRETING THE COEFFICIENTS

```python
# Standerdized coefficients are coefficients of regression where all variables are standardised.
# Odds ratio are what we get after we get exponentials of the coefficients.
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```
